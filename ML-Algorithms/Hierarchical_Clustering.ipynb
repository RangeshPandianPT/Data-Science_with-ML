## Hierarchical Clustering 

**What is Hierarchical Clustering**

- Hierarchical clustering, also known as Hierarchical Cluster Analysis (HCA), is a method of cluster analysis.

- It groups data points based on their similarity.

- The result is a hierarchy of clusters represented in a tree-like structure.

**Key Idea**

- Hierarchical clustering builds a tree structure called a dendrogram.

- The tree shows how clusters are formed or divided step by step.

- Similar data points are grouped closer together in the hierarchy.

**Tree Structure from Data Similarity**

- Clusters are formed based on distance or similarity measures.

- The hierarchy shows relationships between individual points and clusters.

- The height of the tree indicates the level of dissimilarity.

### **Types of Hierarchical Clustering**

**1. Agglomerative Clustering**

- Uses a bottom-up approach.

- Each data point starts as its own cluster.

- The closest clusters are merged step by step.

- Most commonly used hierarchical method.

**2. Divisive Clustering**

- Uses a top-down approach.

- All data points start in a single cluster.

- The cluster is repeatedly split into smaller clusters.

- Less commonly used due to higher computational cost.

**Approaches Summary**

- Bottom-up approach is used in Agglomerative Clustering.

- Top-down approach is used in Divisive Clustering.

**Important Characteristics**

- Does not require the number of clusters in advance.

- Produces a full hierarchy of clusters.

- Once a merge or split is done, it cannot be undone.

**Applications**

- Document and text clustering.

- Customer segmentation.

- Biological and genetic data analysis.

- Image processing and pattern recognition.

## **3. Agglomerative Clustering**

**What is Agglomerative Clustering?**

- Agglomerative Clustering is a type of hierarchical clustering.

- It follows a bottom-up approach.

- Each data point starts as an individual cluster.

- Clusters are gradually merged based on similarity.

**Basic Idea**

- Initially, every data point is its own cluster.

- At each step, the two closest clusters are merged.

- The process continues until a single cluster remains or a stopping condition is met.

**Steps in Agglomerative Clustering**

1. Step 1: Start with Individual Clusters

- Treat each data point as a separate cluster.

- Total number of clusters equals number of data points.

2. Step 2: Compute Distance Matrix

- Calculate distances between all clusters.

- Common distance measures include Euclidean, Manhattan, and Cosine distance.

3. Step 3: Merge Closest Clusters

- Identify the two clusters with the minimum distance.

- Merge them into a single cluster.

4. Step 4: Update Distance Matrix

- Recalculate distances between the new cluster and remaining clusters.

- Distance calculation depends on the chosen linkage method.

5. Step 5: Repeat

- Continue merging clusters step by step.

- Stop when the desired number of clusters is reached or only one cluster remains.

**Linkage Methods**

- Single Linkage: Distance between closest points of two clusters.

- Complete Linkage: Distance between farthest points of two clusters.

- Average Linkage: Average distance between all points of two clusters.

- Wardâ€™s Method: Minimizes variance within clusters.

**Dendrogram**

- A tree-like diagram that shows cluster merging steps.

- Helps decide the optimal number of clusters by cutting the tree at a certain level.

# Hands on Practical Implementation:

## Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

## Step 2: Create a Sample Dataset

```python
data = {
    'Feature_1': [5, 10, 15, 20, 25, 30],
    'Feature_2': [3, 8, 14, 18, 24, 28]
}

df = pd.DataFrame(data)
df
```

## Step 3: Scale the Data

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

## Step 4: Create a Dendrogram

```python
plt.figure(figsize=(10, 6))
linked = linkage(scaled_data, method='ward')

dendrogram(
    linked,
    orientation='top',
    distance_sort='descending',
    show_leaf_counts=True
)

plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

## Step 5: Apply Agglomerative Clustering

```python
hc = AgglomerativeClustering(
    n_clusters=2,
    linkage='ward'
)

clusters = hc.fit_predict(scaled_data)
```

## Step 6: Add Cluster Labels to Dataset

```python
df['Cluster'] = clusters
df
```

## Step 7: Visualize the Clusters

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Feature_1',
    y='Feature_2',
    hue='Cluster',
    data=df,
    palette='coolwarm',
    s=100
)

plt.title('Hierarchical Clustering Result')
plt.show()
```
