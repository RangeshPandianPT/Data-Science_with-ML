# Hypotheses Testing :

- It is a Intermediate Analysis we do , before making a decision

#### **Hypotheses:** Juat an Intuition
- **Null Hypotheses(H0):** Current Process/No Statistically Significant Difference
- **Alternate Hypotheses(Ha):** Alternate Process / Yes,Statistically Siginificant Difference


### Example:

### Scenario:
**Rangesh Launched a new 30-days Python Bootcamp. He want to check whether the average quiz score of his students is different from 75 (Expected passing score)**

### Business Question: 

##### **"Is the Average quiz score of the bootcamp students significantly different from the expected score of 75"**

## Data we Collected: 
- You take a random sample of 50 student scores from the course

## Hypotheses SetUp
- NullHypotheses(H0):μ = 75 (Average score is 75)
- Alternate Hypotheses(Ha): μ != 75 (Average score is not 75)

## Why this test?
To validate your bootcamp's effectiveness.If the average is significantly lower , we will revise the course or instructions.

## Statistical Test

**The test we are conducting to prove the alternative hypothesis should be considered, is what we call a Statistical Test.**
