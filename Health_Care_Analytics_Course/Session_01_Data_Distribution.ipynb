```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python
data = [165, 175, 160, 170, 180, 172, 168, 185, 190, 155]
```

```python
plot.hist(data, bins=5, edgecolor='black')

plot.title("Height Distributions")

plot.xlabel("Height(cms)")

plot.ylabel("Frequency")

plot.show()
```

```python
plot.boxplot(data)

plot.title("Height Distribution")

plot.ylabel("Height(Cms)")

plot.show()
```
