## Level-3: Give Multiple Coditions

1.User should enter the number between 1-100.If it's entered wrong,the program should say,"Invalid Number"

2.If the user guessed number > secret number,then program should say,"You number is greater than the secret number"

3.If the user guessed number < secret number,then program should say,"You number is lesser than the secret number"



```python
print ("Welcome to Number Guessing Game!!!")
print ("Guess the Number between 1 to 100")

secret_number = 56 

for i in range (1,8):
    user_input = int(input("Enter the Number"))
    if user_input < 1 or user_input > 100:
        print("Invalid Number is Guessed!, enter the number between 1-100")
    elif user_input > secret_number:
        print ("You'r number is greater than the secret number!")
    elif user_input < secret_number:
        print("You'r number is lesses than the secret number!")
    else:
        print("You'r guess is correct!")
        break
```

## Level-6: How to handle Invalid Characters or values

```python
import random

print ("Welcome to Number Guessing Game!!!")
print ("Guess the Number between 1 to 100")

secret_number = random.randint(1,100)

for i in range (1,8):
    try: 
        user_input = int(input("Enter the Number"))
        if user_input < 1 or user_input > 100:
            print("Invalid Number is Guessed!, enter the number between 1-100")
        elif user_input > secret_number:
            print ("You'r number is greater than the secret number!")
        elif user_input < secret_number:
            print("You'r number is lesses than the secret number!")
        else:
            print("You'r guess is correct!")
            break
    except ValueError:
        print("Invalid Characters , Not Recognized. Please enter the number between 1-100")

if secret_number == user_input:
    print(f"Congradulations,You have guesses the number {secret_number} inn {i} attemps")
else:
    print("Sorry! All your trails is exceeded. Try Again !")
```
