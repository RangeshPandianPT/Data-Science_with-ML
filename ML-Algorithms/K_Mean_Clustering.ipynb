# Introduction to Clustering

## What is Clustering

- Clustering is the process of grouping objects based on the information present in the data.

- Objects are grouped according to their similarity or relationship with each other.

- It is an unsupervised learning technique, meaning no predefined labels are used.

**1. Main Objective of Clustering**

- Objects within the same group or cluster should be highly similar to each other.

- Objects belonging to different clusters should be significantly different from one another.

- The aim is to maximize intra-cluster similarity and minimize inter-cluster similarity.

**2. Key Characteristics**

- Works on unlabeled data.

- Helps in discovering hidden patterns or structures in data.

- Groups data points based on distance, density, or connectivity measures.

**3. Why Clustering is Used**

- To understand the natural grouping in data.

- To simplify large datasets by summarizing them into clusters.

- To identify patterns, trends, or relationships that are not immediately obvious.

**3. Examples of Applications**

- Customer segmentation in marketing.

- Grouping similar documents or articles.

- Image segmentation.

- Social network analysis.

  
**4. Important Point**

- Clustering does not predict outcomes.

- It focuses on data exploration and structure discovery rather than classification.

# 1. **K Mean Clustering**

## How Does K-Means Work ?

### **Overview**

- K-Means is an unsupervised clustering algorithm.

- It partitions data into K distinct clusters based on similarity.

- Each cluster is represented by the mean of its data points, called the centroid.

### **Step-by-Step Working of K-Means**

#### **Step 1: Choose Number of Clusters (K)**

- Decide the number of clusters K in advance.

- K represents how many groups the data will be divided into.

- Selection of K can be guided using methods like the Elbow Method.

#### **Step 2: Initialization**

- Randomly select K data points as initial centroids.

- These centroids act as the starting points for clusters.

- Poor initialization can affect final clustering results.

#### **Step 3: Cluster Assignment**

- Calculate the distance between each data point and all centroids.

- Assign each data point to the nearest centroid.

- Common distance metric used is Euclidean distance.

#### **Step 4: Move Centroid**

- Recalculate the centroid of each cluster.

- The new centroid is the mean of all data points assigned to that cluster.

- Centroids shift position after reassignment.

#### **Step 5: Optimization**

- Steps of cluster assignment and centroid movement are repeated.

- The algorithm tries to minimize within-cluster variance.

- This process improves cluster compactness.

#### **Step 6: Convergence**

- The algorithm stops when centroids no longer change significantly.

- Or when data point assignments remain unchanged.

- This indicates that an optimal or stable clustering is achieved.

### **Key Characteristics**

- Simple and efficient for large datasets.

- Sensitive to initial centroid selection.

- Works best when clusters are spherical and well separated.

## **Important Note**

- K-Means always converges but not necessarily to the global optimum.

- Results can vary with different initializations.

## **2. Fuzzy C-Means Clustering**

**What is Fuzzy C-Means (FCM)?**

- Fuzzy C-Means is an unsupervised clustering algorithm.

- It is a soft clustering technique.

- Unlike K-Means, a data point can belong to multiple clusters.

- Each data point has a membership value for every cluster.

**Key Concept**

- Membership values range between 0 and 1.

- The sum of membership values of a data point across all clusters equals 1.

- Higher membership value means stronger association with that cluster.

**Difference Between K-Means and Fuzzy C-Means**

- K-Means uses hard clustering, each data point belongs to exactly one cluster.

- Fuzzy C-Means uses soft clustering, data points can partially belong to multiple clusters.

- FCM provides more flexibility for overlapping data.

**How Fuzzy C-Means Works**

- Choose the number of clusters C.

- Initialize membership values randomly.

- Compute cluster centers using weighted average of data points.

- Update membership values based on distance from cluster centers.

- Repeat until membership values and cluster centers stabilize.

**Objective Function**

- FCM minimizes a weighted within-cluster sum of squared errors.

- Weights are based on membership values raised to a fuzziness parameter.

**Fuzziness Parameter (m)**

- Controls the degree of fuzziness in clustering.

- m > 1, commonly m = 2.

- Higher m means more overlap between clusters.

## **3. Agglomerative Clustering**

**What is Agglomerative Clustering?**

- Agglomerative Clustering is a type of hierarchical clustering.

- It follows a bottom-up approach.

- Each data point starts as an individual cluster.

- Clusters are gradually merged based on similarity.

**Basic Idea**

- Initially, every data point is its own cluster.

- At each step, the two closest clusters are merged.

- The process continues until a single cluster remains or a stopping condition is met.

**Steps in Agglomerative Clustering**

1. Step 1: Start with Individual Clusters

- Treat each data point as a separate cluster.

- Total number of clusters equals number of data points.

2. Step 2: Compute Distance Matrix

- Calculate distances between all clusters.

- Common distance measures include Euclidean, Manhattan, and Cosine distance.

3. Step 3: Merge Closest Clusters

- Identify the two clusters with the minimum distance.

- Merge them into a single cluster.

4. Step 4: Update Distance Matrix

- Recalculate distances between the new cluster and remaining clusters.

- Distance calculation depends on the chosen linkage method.

5. Step 5: Repeat

- Continue merging clusters step by step.

- Stop when the desired number of clusters is reached or only one cluster remains.

**Linkage Methods**

- Single Linkage: Distance between closest points of two clusters.

- Complete Linkage: Distance between farthest points of two clusters.

- Average Linkage: Average distance between all points of two clusters.

- Ward’s Method: Minimizes variance within clusters.

**Dendrogram**

- A tree-like diagram that shows cluster merging steps.

- Helps decide the optimal number of clusters by cutting the tree at a certain level.

## 4. Divisive Clustering 

**What is Divisive Clustering?**

- Divisive Clustering is a type of hierarchical clustering.

- It follows a top-down approach.

- All data points start in a single cluster.

- The cluster is repeatedly divided into smaller clusters.

**Basic Idea**

- Begin with one large cluster containing all data points.

- At each step, split one cluster into two or more sub-clusters.

- Continue dividing until each data point becomes its own cluster or a stopping criterion is met.

**Steps in Divisive Clustering**

1. Step 1: Start with One Cluster

- Place all data points into a single cluster.

- This represents the root of the hierarchy.

2. Step 2: Select Cluster to Split

- Choose the cluster with the highest dissimilarity or variance.

- This cluster is the best candidate for division.

3. Step 3: Split the Cluster

- Divide the selected cluster into two smaller clusters.

- Splitting can be done using algorithms like K-Means with K = 2.

4. Step 4: Evaluate the Split

- Measure the quality of the split using distance or variance metrics.

- Ensure clusters are more compact than before.

5. Step 5: Repeat

- Continue splitting clusters recursively.

- Stop when the desired number of clusters is reached or clusters cannot be meaningfully divided.

**Dendrogram**

- The hierarchical structure can be represented using a dendrogram.

- Divisive dendrogram is read from top to bottom

## 5. Mean Shift Clustering

**What is Mean Shift Clustering?**

- Mean Shift is an unsupervised clustering algorithm.

- It is a centroid-based and density-based clustering technique.

- The main goal is to discover dense regions or blobs in the data.

**Core Idea**

- The algorithm identifies areas in the feature space where data points are highly concentrated.

- Each cluster corresponds to a mode or peak in the data density.

**How Mean Shift Works?**

1. Step 1: Define a Window (Bandwidth)

- A circular or kernel window is defined around each data point.

- Bandwidth controls the size of the region considered.

- Choice of bandwidth significantly affects the results.

2. Step 2: Compute Mean of Points

- For each window, compute the mean of all data points inside the region.

- This mean represents a candidate centroid.

3. Step 3: Shift the Window

- Move the window center to the newly computed mean.

- This step shifts the window toward higher data density.

4. Step 4: Repeat

- Continue shifting until the window converges to a stable position.

- Stable positions represent density peaks.

5. Step 5: Post-Processing

- Candidate centroids that are very close to each other are merged.

- The remaining centroids form the final cluster centers.

**Key Characteristics**

- Does not require specifying the number of clusters in advance.

- Automatically finds the number of clusters.

- Clusters can have arbitrary shapes.
