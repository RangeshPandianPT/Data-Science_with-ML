```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Data
data = {
    "name": ["Max Aarons", "Joshua Acheampong", "Tyler Adams", "Tosin Adarabioyo", "Simon Adingra"],
    "team": ["Bournemouth", "Chelsea", "Bournemouth", "Chelsea", "Brighton"],
    "minutes": [86, 170, 1875, 1319, 1052],
    "goals": [0, 0, 0, 1, 2],
    "assists": [0, 0, 3, 1, 2],
    "expected_goals": [0.0, 0.2, 1.6, 0.9, 2.4],
    "progressive_carries": [1, 0, 13, 5, 50],
    "progressive_passes": [8, 8, 71, 39, 18],
    "yellow": [0, 1, 7, 3, 0],
    "red": [0, 0, 0, 0, 0]
}

df = pd.DataFrame(data)

# Features to normalize
features = [
    "minutes", "goals", "assists", "expected_goals",
    "progressive_carries", "progressive_passes",
    "yellow", "red"
]

# Normalize features
scaler = MinMaxScaler()
df_norm = df.copy()
df_norm[features] = scaler.fit_transform(df[features])

# Weights
weights = {
    "goals": 0.25,
    "assists": 0.15,
    "expected_goals": 0.20,
    "progressive_carries": 0.10,
    "progressive_passes": 0.10,
    "minutes": 0.10,
    "yellow": -0.05,
    "red": -0.05
}

# Weighted score calculation
df_norm["player_score"] = sum(
    df_norm[feature] * weight
    for feature, weight in weights.items()
)

# Aggregate by team
team_scores = df_norm.groupby("team")["player_score"].sum().sort_values(ascending=False)

print("Normalized Team Scores:")
print(team_scores)

# Final prediction
winner = team_scores.idxmax()
print("\nPredicted Winning Team:", winner)
```
