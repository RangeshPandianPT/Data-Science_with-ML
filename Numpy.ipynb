 # Reshaping and Array Conversions

```python
import numpy as np

n10 = np.array([1,2,3,4,5,6,7])
n10_padded = np.append(n10, 0)   # now size = 8
n12 = n10_padded.reshape(2,4)
print(n12)
```

```python
n13 = np.arange(4)
print(n13)
n14 = np.arange(5,9)
print(n14)
```

```python
n15 = np.concat((n13,n14))
print(n15)
```

```python
#Converting 1D array to 2D array and so on ....
n16 = np.arange(8)
print(n16)
n17 = n16[np.newaxis,:]
print(n17)
print(n17.shape)
n18 = n17[np.newaxis,:]
print(n18)
print(n18.shape)
```

```python
n19 = n16[:,np.newaxis]
print(n19)
print(n19.shape)
```

```python
n20 = np.expand_dims(n16 , axis = 1)
print(n20)
```

```python
n20 = np.expand_dims(n16 ,axis = 0)
print(n20)
```

# Copy

### Shallow copy

```python
n24 = np.arange(20)
n25 = np.arange(20)
```

```python
n24 = n25
```

```python
print(n25)
```

```python
n24[0] = 100
print(n25)  # Memory is changed to n25 also when we change in n24
```

### Deep copy

```python
n26 = n24.copy()
n24[1] = 200
print(n24)
print(n26)
```
