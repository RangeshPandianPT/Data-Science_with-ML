```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# Select ONLY numeric columns
numeric_data = titanic.select_dtypes(include=["number"])

sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```

## 3. Multivariate Analysis

```python

sns.catplot(x="class", hue="sex", col="survived",
            data=titanic, kind="count")
plt.show()
```

```python
sns.scatterplot(x="age", y="fare", hue="survived", style="sex", data=titanic)
plt.title("Age vs Fare colored by Survival & styled by Gender")
plt.show()
```

```python
sns.boxplot(x="class", y="age", hue="survived", data=titanic)
plt.title("Age Distribution across Class & Survival")
plt.show()
```

```python
sns.violinplot(x="class", y="fare", hue="sex", split=True, data=titanic)
plt.title("Fare Distribution by Class & Gender")
plt.show()
```

```python
numeric = titanic.select_dtypes(include="number")
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
