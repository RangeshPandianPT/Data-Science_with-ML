```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("player_stats.csv")
df
```

```python
features = [
    "minutes",
    "goals",
    "assists",
    "expected_goals",
    "progressive_carries",
    "progressive_passes",
    "yellow",
    "red"
]
```

```python
scaler = MinMaxScaler()

df_norm = df.copy()
df_norm[features] = scaler.fit_transform(df[features])
```

```python
weights = {
    "goals": 0.25,
    "assists": 0.15,
    "expected_goals": 0.20,
    "progressive_carries": 0.10,
    "progressive_passes": 0.10,
    "minutes": 0.10,
    "yellow": -0.05,
    "red": -0.05
}
```

```python
df_norm["player_score"] = 0

for feature, weight in weights.items():
    df_norm["player_score"] += df_norm[feature] * weight
```

```python
team_scores = (
    df_norm
    .groupby("team")["player_score"]
    .sum()
    .sort_values(ascending=False)
)

print("Team Strength Scores (Normalized):")
print(team_scores)
```

```python
predicted_winner = team_scores.idxmax()
print("\nPredicted Winning Team:", predicted_winner)
```
