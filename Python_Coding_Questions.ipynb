## 2. Create a list of numbers and print only even number using a for loop.

```python
numbers = range(1,20)
even_no = []

for i in numbers:
    if i%2 == 0:
        even_no.append(i)
print(even_no)
```

## 3. Write a program that handles division by zero using try-except.

```python
get_num = int(input("Enter a number (Numerator):"))
get_den = int(input("Enter a number (Denominator):"))

try:
    result = (get_num/get_den)
    print(result)
except ZeroDivisionError:
    print("Invalid Entry !")

except ValueError:
    print("Invalid Entry !")
    
finally:
    print("Code successfully ended")
```

### 4. Given a string , count how many vowels it has

```python
user_input = input("Enter any word or sentence")
count = 0

for i in user_input.lower():
    if i in "aeiou":
        print(i)
        count += 1

print("Total no of Vowels are:",count)
```

### 5. Reverse a list without using built-in-function

```python
original_list = [1,2,3,4,5,6,7]
updated_list = []
i = len(original_list) - 1
while i>=0:
    updated_list.append(original_list[i])
    i -=1
    
updated_list
    
```
