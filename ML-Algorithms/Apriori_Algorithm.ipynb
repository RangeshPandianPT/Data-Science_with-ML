## Association Rule Mining 

### **What is Association Rule Mining?**

- Association Rule Mining is a data mining technique.

- It discovers relationships or patterns between items in large datasets.

- Commonly used in market basket analysis.

**Rule Format**

- Rules are written in the form:
A → B

- Meaning: If A occurs, then B is likely to occur.

**Key Components**

- Antecedent (A): The item or condition on the left side (IF part).

- Consequent (B): The item or outcome on the right side (THEN part).

**Example**

- {Bread} → {Butter}

- If a customer buys bread, they are likely to buy butter.

**Important Measures**

- Support: Frequency of A and B occurring together.

- Confidence: Probability of B occurring when A occurs.

- Lift: Strength of the rule compared to random occurrence.

**Characteristics**

- Works on transactional data.

- Does not require labeled data.

- Helps uncover hidden buying patterns.

**Applications**

- Market basket analysis.

- Recommendation systems.

- Cross-selling strategies.

- Web usage mining.

## Association Rule Mining 

**Rule Representation**

- Association rules are written as: A → B

- Meaning: If item A occurs, item B is likely to occur.

### Support

- Support measures how frequently items A and B occur together.

- It indicates the importance of a rule in the dataset.

**Formula**

- **Support(A → B) = freq(A, B) / N**


**Where:**

- freq(A, B) = number of transactions containing both A and B

- N = total number of transactions

**Interpretation**

- Higher support means the rule appears more often.

- Low support rules may not be useful.

### Confidence

- Confidence measures how often B occurs when A has already occurred.

- It indicates the reliability of the rule.

**Formula**

- **Confidence(A → B) = freq(A, B) / freq(A)**


**Interpretation**

- High confidence means strong implication.

- It does not consider how common B is overall.

### Lift

- Lift measures the strength of a rule compared to random chance.

- It shows whether A and B are dependent or independent.

**Formula**

- **Lift(A → B) = Support(A, B) / (Support(A) × Support(B))**


**Interpretation**

- Lift > 1: Positive association between A and B

- Lift = 1: A and B are independent

- Lift < 1: Negative association

**Summary Table**

- Support: Frequency of A and B together

- Confidence: Reliability of rule A → B

- Lift: Strength and usefulness of the rule

# Hands on Practical 

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df = pd.read_csv('OnlineRetail.csv', encoding='latin1')
df.head()
```

```python
basket = (
    df[df['Country'] == "France"]
    .groupby(['InvoiceNo', 'Description'])['Quantity']
    .sum()
    .unstack()
    .reset_index()
    .fillna(0)
    .set_index('InvoiceNo')
)

basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    else:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(
    basket_sets,
    min_support=0.07,
    use_colnames=True
)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)

rules.head()
```
