## Note 

- **Importance of Self and Init comes into the picture**

```python
class SecureUser:
    def __init__(self ,name ,atm_pin):
        self.name = name
        self.__atm_pin = atm_pin

    def get_pin(self , attempt):
        return attempt == self.atm_pin

    def deposite_cash(self):
        pass

    def cash_withdraw(self):
        pass
```

```python
user_data =SecureUser("Rangesh",2828)
```
