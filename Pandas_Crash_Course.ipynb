## 1.3 Give me the list of Top selling Items

Done by chain of function methods

```python
order_details.groupby(by="item_name")["quantity"].sum() #Selling items with respective to quantity
```

```python
order_details.groupby(by="item_name")["quantity"].sum().sort_values(ascending = False) #Arranging the quantities in a decending order 
```

```python
order_details.groupby(by="item_name")["quantity"].sum().sort_values(ascending = False).head(10)  #Finding the Top 10 selling items detail 
```

## 1.3 Give me the list of Least 5 Selling Items

```python
order_details.groupby(by="item_name")["quantity"].sum().sort_values(ascending = True).head(10)  #Finding the Top 10 least selling items 
```
