```python
type(student_marks)
```

```python
items_to_buy = ["Apple","Orange","Graps","Cherry"]
items_to_buy
```

```python
len(items_to_buy)
```

```python
items_to_buy[0] #Indexing
```

```python
items_to_buy[-2] #negetive Indexing
```

## 2. Sets

- 1D
- Heterogenous type
- No duplicate Values
- Indexing and Slicing is not possible
- Mutable 

```python
empty_cart = set() #empty set 
```

```python
type(empty_cart)
```

```python
items_to_buy_2 = {"Apple","Orange","Graps","Cherry" }
items_to_buy_2
```

```python
type(items_to_buy_2)
```

```python
items_to_buy_2.add("watermelon")
```

```python
items_to_buy_2
```

```python
items_to_buy_2.add(69.5)
```

```python
items_to_buy_2
```

```python
items_to_buy_2.discard("watermelon")
```

```python
items_to_buy_2
```

```python
items_to_buy_2.copy()
```

```python
items_to_buy_2.pop() #Stack Concept
```
