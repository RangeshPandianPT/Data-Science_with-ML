# Guess the number Game level-wise

## 1. Print if the user guessed number is right or not 

```python
secret_number = 56

user_input = int(input("Enter your guess: "))

if secret_number == user_input:
    print("Congratulations! Your guess is right!")
else:
    print("Sorry. Your guess is wrong")
```

## Level-2:Print if the user guessed number is right or not but with max of 7 trails

#### 2.1 Using For Loop - end interation number

```python
secret_number = 56

for i in range (0,7):
    user_input = int(input("Enter your guess"))

    if secret_number == user_input:
        print("Congradulation! You're guess is right!")
        break
    else:
        print("Sorry , Your guess is wrong , try again")
         
```

#### 2.2 While Loop- When end number is not defined and the instructions has to executed infinite times 

```python
secret_number = 56 

while True:
    user_input = int(input("Enter your guess"))

    if secret_number == user_input:
        print ("Congradulation! You're guess is right!")
        break 
    else:
        print("Sorry, Your guess is wrong")
        
```
