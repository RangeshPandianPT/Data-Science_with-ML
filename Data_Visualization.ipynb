```python
titanic_data = sns.load_dataset(name ="titanic" )
titanic_data
```

```python
sns.countplot(x="survived", data=titanic_data)
plt.title("Survival Count")
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(titanic_data)
df.to_csv("Titanic.csv", index=False) #From here importing the csv file out
```

```python
sns.countplot(x="sex", data=titanic_data)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x="class", data=titanic_data)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
titanic_data["embarked"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Embarked Port Distribution")
plt.show()
```

```python
sns.histplot(titanic_data["age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=titanic_data["age"])
plt.title("Age Boxplot")
plt.show()
```
